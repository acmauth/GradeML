```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("csd_2021.csv")
```

```python
courses = data.iloc[:,10:]
courses
```

```python
import json
from pandas.io.json import json_normalize

with open("courses_ids_600000014.json","r") as json_file:
     json_file = json.load(json_file)
coded_courses = json_normalize(json_file['courses'])
del coded_courses['ccoursecode']
course_dict = coded_courses.set_index('coursecode')['courseId'].to_dict()

for key in course_dict.keys():
    courses.rename(columns = {key:str(course_dict[key]).strip()}, inplace=True)
display(courses)
```

```python
courses_metrics = dict()
for course in courses.columns:
    
    current_course = courses[course]
    current_course = (current_course[current_course >= 0]).to_numpy()
    if len(current_course) > 0:
        courses_metrics[course] = dict()
        # print(current_course)
        average = np.average(current_course)
        print(average)
        courses_metrics[course]['average'] = float(average)
        if average < 5.0:
            difficulty =  "VERY HARD"
        elif average <= 6:
            difficulty = "HARD"
        elif average <= 8:
            difficulty =  "MEDIUM"
        else:
            difficulty = "EASY"
        print(difficulty)
        courses_metrics[course]['difficulty'] = difficulty
        print(len(current_course))
        courses_metrics[course]['enrolled'] = len(current_course)

        bins, edges = np.histogram(current_course)
        print(bins)
        courses_metrics[course]['histogram'] = bins.tolist()
```

```python
print(len(courses_metrics))
```

```python
import json
with open('course_metrics.json', 'w') as fp:
    json.dump(courses_metrics, fp)
```
